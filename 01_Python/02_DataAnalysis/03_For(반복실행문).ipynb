### for문
* 반복 실행문
* for 변수 in 반복실행사용데이터:
* ....실행문1
* ....실행문2
* 실행문4
* 반복실행에 사용할 데이터를 list, range, dictionary.keys() 형태로 제공
* 해당 데이터를 하나씩 사용하여 모두 소진할때까지 반복 실행

```python
# i 는 반복실행을 제어하는 변수
# range(0, 10) : 반복횟수를 정해주는 범위내 숫자 발생 함수
# 반복 실행의 시작 명령부터 끝 명령까지를 들여쓰기를 통해 표시해줍니다.
for i in range(0, 10):
    print("겁나 쉬운 파이썬 프로그래밍")
```

```python
# range(시작숫자, 끝숫자+1)
# 시작 숫자 부터 끝숫자까지의 정수를 발생해 주는 함수입니다
# 발생된 숫자들은 차례로 i변수에 저장되면서 반복실행이 진행됩니다.
# 아래 예제는 1부터 9까지 정수를 발행하고 그 갯수 만큼 반복실행하는 예
# i 변수가 반복 실행마다 저장하는 값을 출력해봅니다
for i in range(1, 10):
    print(i, end=" ")
# print 명령에 줄바끔 기능을 없애려면  end="" 또는 end=" " 를 씁니다
```

```python
# 10번의 반복을 위해서 반드시 range(0, 10) 이 필요한것은 아닙니다
for i in range(10, 20):
    print(i, end=" ")
print()
for i in range(100, 110):
    print(i, end=" ")
```

```python
# range 함수는 리스트의 슬라이싱처럼 증감량이 존재하여 그 크기를 조절할수
# 있습니다
for i in range(10, 0, -1):
    print(i, end=" ")
```

```python
# range 를 이용한 다양한 반복실행의 예
# 0 부터 99까지의 정수 출력
for i in range(0,100):
    print(i, end=" ")
print("\n")
# 1부터 100 사이의 짝수 출력 1
for i in range(2, 100, 2):
    print(i, end=" ")    
```

#### 연습문제2
* 정수를 입력 받고, 입력 받은 정수의 factorial 을 계산하여 출력하세요
* (입력)정수입력 : 6
* (출력)6! = 1x2x3x4x5x6 = 720

```python
n = int( input("정수 입력 : ") )
fact = 1
print( n, "! = ", end = "")
for i in range(1, n+1):
    print(i, "x", end=" ")
    fact = fact * i
print("\b\b = ", fact)
```

#### 연습문제3
* 년, 월, 일을 입력 받아서 요일을 출력하세요
* 년도에 해당하는 계산은 입력한 년도의 전년도에 365를 곱하고,
* 1부터 전년도까지 윤년인지를 모두 검사해서 윤년이 나오면 1증가하는 반복문을 사용하세요

```python
year = int( input("년도 입력:"))
month = int( input("월 입력 : ") )
day = int( input("일 입력 : ") )

# 입력한 년도의 전년도까지 날짜수를 365 를 곱해서 계산
total = 365 * (year-1)

# 입력한 년도의 전년도까지 윤년이 몇번있었는지 검사해서 1씩 덧샘합니다
for i in range(1, year):
    if (i%4==0) and (i%100!=0) or (i%400==0):
        total = total + 1
        
# 입력한년도 윤년인지 검사한 후 feb 변수에 28 또는 29  를 저장한 후
# 2월의 날짜수로 활용합니다
if (year%4==0) and (year%100!=0) or (year%400==0):
    feb = 29
else:
    feb = 28

# 위 상황에 맞게 아래 if 문을 수정합니다.
if month==1: total = total
elif month==2: total = total + 31
elif month==3: total = total + 31+feb
elif month==4: total = total + 31+feb+31
elif month==5: total = total + 31+feb+31+30
elif month==6: total = total + 31+feb+31+30+31
elif month==7: total = total + 31+feb+31+30+31+30
elif month==8: total = total + 31+feb+31+30+31+30+31
elif month==9: total = total + 31+feb+31+30+31+30+31+31
elif month==10: total = total + 31+feb+31+30+31+30+31+31+30
elif month==11: total = total + 31+feb+31+30+31+30+31+31+30+31
elif month==12: total = total + 31+feb+31+30+31+30+31+31+30+31+30
# 조건에 걸린 명령이 한개만 있다면 위와 같이 if 또는 elif 와 나란히 표기가
# 가능합니다
total = total + day
n = total % 7
if n==1: print("월요일입니다")
elif n==2: print("화요일입니다")
elif n==3: print("수요일입니다")
elif n==4: print("목요일입니다")
elif n==5: print("금요일입니다")
elif n==6: print("토요일입니다")
elif n==0: print("일요일입니다")
```

```python
year = int( input("년도 입력:"))
month = int( input("월 입력 : ") )
day = int( input("일 입력 : ") )
total = 365 * (year-1)
for i in range(1, year):
    if (i%4==0) and (i%100!=0) or (i%400==0):
        total = total + 1
mdays = [0, 31, 28, 31 ,30 ,31 ,30 ,31 ,31 ,30, 31 ,30 ,31]
if (year%4==0) and (year%100!=0) or (year%400==0):
    mdays[2] = 29
for i in range(1, month):
    total = total + mdays[i]
total = total + day
n = total % 7
if n==1: print("월요일입니다")
elif n==2: print("화요일입니다")
elif n==3: print("수요일입니다")
elif n==4: print("목요일입니다")
elif n==5: print("금요일입니다")
elif n==6: print("토요일입니다")
elif n==0: print("일요일입니다")
```

#### 정수를 입력받아서 소수인지 아닌지를 출력하세요
* 소수(prime number)를 판단하는 방법
    * n 에 저장된 값이 소수인지 아닌지 판단하려면, 1부터 n 까지의 모든 숫자로 n 을 나눕니다
    * 나눈 나머지가 0인 경우가 몇번이었는지 count 하고 count 값이 2이면 소수로 출력합니다
    * count가 2라는건 1과 자기자신(n) 으로 나눴을때만 나머지가 0이었다는 뜻

```python
n = int(input("정수 입력 : "))
count = 0
for i in range(1, n+1):
    if n % i == 0 :
        count = count+1
if count==2:
    print("소수입니다")
else :
    print("소수가 아닙니다")
```
